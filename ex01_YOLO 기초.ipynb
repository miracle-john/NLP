# 🤖 **1. 객체 탐지 (Object Detection)**

컴퓨터 비전 기술 중 하나로, 이미지나 영상 속에서 **<U>특정 객체의 종류(what)와 위치(where)를 찾아내는 기술</U>**, 단순히 이미지를 분류하는 것을 넘어, 객체 주변에 경계 상자(bounding box)를 그리고 어떤 객체인지 레이블(label)을 붙여주는 것이 핵심!

## 👉 1-1. 이미지 분류와의 차이점

- **Classification(분류):** 여러 물체에 대해 어떤 물체인지 <U>**클래스를 분류**</U>하는 문제
- **Localization(위치 파악):** 물체가 어디 있는지 박스를 통해 <U>**위치 정보**</U>를 나타내는 문제
- **Object Detection(객체탐지):** Classification과 Localization을 <U>**동시에 진행**</U>하는 것

## 👉 2-2. YOLO 이론

### 1) YOLO 모델 이전의 객체 탐지

<img src="https://drive.google.com/uc?export=view&id=1bDzNBBX-eIUrXFIdi7vlTW5He_kL_hUp">


- R-CNN은 **Regions with CNN features**의 약자로 이미지 속에서 객체가 있을 만한 <U>**영역(Regions)**</U>을 먼저 찾고, 그 영역들을 <U>**CNN(딥러닝)**</U>을 이용해 분석하는 방식

- 작동 단계
  - 1. 후보 영역 제안 (Region Proposal) : Selective Search, Sliding window
  - 2. 이미지 변형 (후보 영역 이미지를 CNN 모델 입력사이즈에 맞게 변형
  - 3. CNN 특징 추출
  - 4. 분류 및 위치 보정

- 장점 : 딥러닝을 적용하여 월등히 높은 정확도를 달성함
- 단점 : 치명적으로 느린 속도, 복잡한 구조(학습과정이 복잡하고 비효율적임)

### 2) YOLO 모델 등장

YOLO는 **You Only Look Once**의 약자로, '후보 영역 제안'과 '분류'를 나누지 않고, <U>이미지 전체</U>를 단 한번만 보고 객체의 <U>위치와 종류를 동시에 예측</U>하는 <U>**1-Stage Detector**</U> 방식

<img src="https://drive.google.com/uc?export=view&id=1IAbW4L9bTbPhgb8DuuKsaKqTTAwYoEeh">


- 작동 단계
  - 1. 이미지 그리드 분할 : S*S 형태의 격자(Grid)로 나눔
  - 2. 단일 네트워크로 동시 예측 : 객체의 Bounding Box와 해당 객체가 무엇일지 동시에 예측
  - 3. 최종 결과 필터링

- 장점 : 압도적으로 빠른속도, 이미지 전체를 보기 때문에 맥락을 이해하여 배경 오류가 적음, 간결한 구조
- 단점 : 상대적으로 낮은 정확도


### 3) YOLO 모델의 원리

<img src="https://drive.google.com/uc?export=view&id=17cvLgkxJlDtulQ28wo8kjOXS-N5r5duS">


> **1단계 - 그리드 분할 :**  입력된 이미지를 S x S 크기의 격자(Grid)로 나눕니다.
>
> **2단계 - 동시 예측 :**  각 격자 셀이 위치와 클래스 정보를 동시에 예측합니다.
> - **Bounding Box 예측:** 객체의 위치(x,y), 크기(w,h), 객체 존재 신뢰도(confidence)를 예측합니다.
> - **클래스 확률 예측:** 해당 셀에 객체가 있을 경우, 어떤 종류(class)일지에 대한 확률을 예측합니다.
>
> **3단계 - 결과 종합 :**  모든 셀의 예측 결과를 하나로 모아 수많은 후보 상자를 생성합니다.
>
> **4단계 - 후처리 (Post-processing) :**  신뢰도가 낮은 상자를 제거하고, 중복된 상자들 중 가장 점수가 높은 하나만 남깁니다 (NMS, Non-Maximum Suppression).

## 👉 2-3. YOLO 모델 평가지표

### YOLO 모델 평가지표 : 모델의 성능을 다양한 관점에서 평가

| 지표                           | 설명                                                                                                                       | 특징 / 활용                                                         |
|--------------------------------|----------------------------------------------------------------------------------------------------------------------------|----------------------------------------------------------------------|
| **IoU (Intersection over Union)**  | - 예측 박스와 실제 박스의 겹치는 영역 비율<br>- 0~1 범위, 1은 완벽 일치                                                    | - 객체 탐지의 기본 기준<br>- mAP, AP50/75 계산 시 임계값 설정에 사용됨  |
| **Precision(정밀도) / Recall(재현율) / F1**    | - **Precision:** 예측한 객체 중 실제 객체 비율<br>- **Recall:** 실제 객체 중 예측 성공 비율<br>- **F1:** 둘의 조화 평균       | - 세밀한 모델 성능 분석에 적합<br>- 클래스 불균형 상황에서 유용         |
| **mAP (mean Average Precision)**  | - 여러 클래스의 AP(Average Precision) 평균값<br>- IoU 임계값(예: 0.5, 0.75 등)에 따라 평가                                    | - 모델 전반의 정확도를 종합적으로 평가<br>- 객체 탐지 분야의 대표 지표     |
| **AP50 / AP75**                | - **AP50:** IoU 0.5 기준 AP<br>- **AP75:** IoU 0.75 기준, 더 엄격한 평가                                                     | - IoU 기준에 따라 모델의 예측 정확도 세분화<br>- 프로토타입 성능 비교에 활용 |
| **FPS (Frames Per Second)**    | - 초당 처리 가능한 이미지(프레임) 수                                                                                        | - 실시간 응용(자율주행, CCTV 등)에서 중요<br>- 속도 평가의 주요 지표       |
| **Inference Time / Latency**   | - 단일 이미지 처리 시간 (밀리초 단위)                                                                                        | - 시스템 반응 속도 파악<br>- 배포 환경 및 하드웨어 최적화 판단에 활용       |


> - 전이학습 후 모델의 정확도를 전반적으로 확인하려면 **mAP**가 가장 대표적 (전체적인 정확도)
> - 특정 IoU 임계값에서의 성능을 더 자세히 보려면 **AP50** 또는 **AP75** 활용 (객체의 위치 탐색의 정밀도)
> - 클래스 불균형 여부나 세밀한 오탐/미탐을 파악하려면 **Precision/Recall/F1** 지표 병행 (각 클래스에 대한 성능 세밀 분석)
> - 공간적 예측의 정확도는 **IoU**로 판단(예측 박스), 속도는 **FPS/Inference Time**을 확인(실시간 처리 능력)

````python
import shutil
import os

# food_list와 food_file_names_list가 정의되어 있다고 가정합니다.
# zip을 사용하여 음식 종류와 해당 파일 리스트를 하나씩 가져옵니다.
for food, img_list in zip(food_list[1:], food_file_names_list):

    # 1. Train set (0~640)
    for img_name in img_list[0:640]:
        src = f'./food/{food}/{img_name}'
        dst = f'./data/food_for_yolo/train/{food}/{img_name}'
        if os.path.exists(src):
            shutil.move(src, dst)

    # 2. Val set (640~800)
    for img_name in img_list[640:800]:
        src = f'./food/{food}/{img_name}'
        dst = f'./data/food_for_yolo/val/{food}/{img_name}'
        if os.path.exists(src):
            shutil.move(src, dst)

    # 3. Test set (800~)
    for img_name in img_list[800:]:
        src = f'./food/{food}/{img_name}'
        dst = f'./data/food_for_yolo/test/{food}/{img_name}'
        if os.path.exists(src):
            shutil.move(src, dst)
````

````python
# 1. 파일 이름 리스트들을 먼저 정의해야 합니다 (셀 [88]에서 수행한 작업)
# 만약 이미 정의했다면 이 부분은 건너뛰어도 됩니다.
gyoza_file_names_list = os.listdir('./food/gyoza')
bibimbapa_file_names_list = os.listdir('./food/bibimbap')
caesar_salad_file_names = os.listdir('./food/caesar_salad')
apple_pie_file_names = os.listdir('./food/apple_pie')
takoyaki_file_names = os.listdir('./food/takoyaki')

# 2. 리스트를 하나로 묶어주는 변수 정의 (현재 누락된 부분)
food_file_names_list = [
    gyoza_file_names_list,
    bibimbapa_file_names_list,
    caesar_salad_file_names,
    apple_pie_file_names,
    takoyaki_file_names
]

# 3. food_list도 정의되어 있는지 확인
food_list = ['gyoza', 'bibimbap', 'caesar_salad', 'apple_pie', 'takoyaki']
````

````python
# 나머지 음식 이미지 파일 옮기기 gyoza, bibimbap, caesar_salad, apple_pie, takoyaki
food_file_names_list = [gyoza_file_names_list, bibimbapa_file_names_list, caesar_salad_file_names, apple_pie_file_names, takoyaki_file_names]

for food, img_list in zip(food_list[1:], food_file_names_list):
  for img_name in img_list[0:640] :
   shutil.move('./food/'+food+'/'+img_name,
                  './food_for_yolo/train/'+food+'/'+img_name)

   for img_name in img_list[640:800]:
    shutil.move('./food/gyoza/'+food+'/'+img_name,
                './food_for_yolo/train/gyoza/'+food+'/'+img_name)

   for img_name in img_list[800:]:
     shutil.move('./food/gyoza/'+food+'/'+img_name,
                './food_for_yolo/train/gyoza/'+food+'/'+img_name)
````

````python
import shutil
import os

# 1. 대상 폴더들이 제대로 만들어져 있는지 확인
for folder in ['train', 'val', 'test']:
    for food in ['gyoza', 'bibimbap', 'caesar_salad', 'apple_pie', 'takoyaki']:
        path = f'./data/food_for_yolo/{folder}/{food}'
        if not os.path.exists(path):
            os.makedirs(path)

# 2. 파일 이동 수행 (파일이 존재할 때만 이동)
for food, img_list in zip(food_list, food_file_names_list):
    for i, img_name in enumerate(img_list):
        src = os.path.join('./food', food, img_name)

        # 0~640: train, 640~800: val, 800~: test
        if i < 640: folder = 'train'
        elif i < 800: folder = 'val'
        else: folder = 'test'

        dst = os.path.join('./data/food_for_yolo', folder, food, img_name)

        if os.path.exists(src):
            shutil.move(src, dst)
        # 파일이 없으면 이미 옮겨졌으므로 무시함
````
